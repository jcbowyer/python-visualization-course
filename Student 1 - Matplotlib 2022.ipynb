Matplotlib has three interfaces (ways to write code)

```python
# Showing subtle different between pyploy and OO approach
x = np.linspace(0, 10, 1000)

# pyplot approach

plt.plot(x, x + 0, linestyle = 'solid')  # plot creates a line plot
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# Object-oriented approach

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1) # the same as fig.add_subplot(111)

ax1.plot(x, x + 0, linestyle='solid')
ax1.plot(x, x + 1, linestyle='dashed')
ax1.plot(x, x + 2, linestyle='dashdot')
ax1.plot(x, x + 3, linestyle='dotted')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data = {
    'ME': 109, 'NH': 103, 'VT': 112, 'MA': 112,
    'CT': 100, 'RI': 103, 'NY': 137, 'PA': 125,
    'OH': 104
}
states = list(data.keys())
values = list(data.values())

# Mean for annotating
state_mean = np.mean(values)

# Creating horizontal bar chart
plt.figure(figsize=(9, 3))  # Width: 9, Height: 3, approximately 3x width to height ratio
plt.barh(states, values, color='green')
plt.xlim(80, 140)  # X-axis limit from 80 to 140
plt.axvline(state_mean, color='red', linewidth=1, linestyle='--')
plt.title('Horizontal Bar Chart of Food Illness')

# Adding a label to the mean line
plt.annotate(f'Mean: {state_mean:.1f}', xy=(state_mean, 0), xytext=(state_mean + 5, 0),
             arrowprops=dict(facecolor='black', shrink=0.05),
             va='center')

plt.show()
```

## add_axes() vs. add_subplot()

The difference between fig.add_axes() and fig.add_subplot() is the mechanisms used.  For add_axes() a list is passd in specifiying the the position of the axes (left, botton,width, height).  This means that the axes object is positioned in **absolute coordinates**.

The add_subplot() function does not permit explicit positioning.  Rather, the axes is positioned according to a subplot grid.

In most cases, add_subplot() is used.  In cases where positioning matters, add_axes() is used. fig =

```python

import matplotlib.pyplot as plt
import numpy as np

# Data
names = ['Georgia', 'Denver', 'Phoenix']
values = [1000, 900, 1200]
dates = ['1/1/2021', '1/2/2021', '1/3/2021']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot (bar chart)
axs[0, 0].bar(names, values, color='blue')
axs[0, 0].set_title("Bar Chart")

# Second subplot (horizontal bar chart)
axs[0, 1].barh(range(len(names)), values, color='blue')
axs[0, 1].set_yticks(range(len(names)))
axs[0, 1].set_yticklabels(names)
axs[0, 1].set_title("Horizontal Bar Chart")

# Third subplot (line plot)
x = np.linspace(0, 1, len(values))
y = np.linspace(0, 5, len(values))
axs[1, 0].plot(x, y, color='blue', marker='s')
axs[1, 0].set_title("Line Plot")

# Fourth subplot (empty plot for spacing)
axs[1, 1].axis('off')

# Display the plot
plt.tight_layout()
plt.show()

```
